```python
import numpy as np
```

```python
def load_data():
  data=np.loadtxt("restaurent_profit.txt",delimiter=',')
  x_train=data[:,0]
  y_train=data[:,1]

  return x_train,y_train
```

```python
load_data()
```

```python
#Cost function
def compute_cost(x,y,w,b):
  m=len(x)
  cost=0
  for i in range(m):
    cost +=(w*x[i]+b - y[i])**2
  cost=cost/(2*m)
  return cost

```

```python
#Comput_gradient
def compute_gradient(x,y,w,b):
  m=len(x)
  dw=np.sum((w*x+b-y)*x)/m
  dp=np.sum(w*x+b-y)/m
  return dw,dp
```

```python
def gradient_descent( x, y, w_init, b_init, alpha, num_iters):
    w = w_init
    b = b_init
    J_history = []

    for i in range(num_iters):
        dw, db = compute_gradient(x, y, w, b)
        w -= alpha * dw
        b -= alpha * db
        J_history.append(compute_cost(x, y, w, b))

    return w, b, J_history
```

```python
#Train the model
w_init = -100
b_init = -300
alpha = 0.01
num_iters = 7000
x_train,y_train=load_data()
w, b, J_history = gradient_descent(x_train,y_train, w_init, b_init, alpha, num_iters)
print(f'Trained parameters: w = {w}, b = {b}')
```

```python
import matplotlib.pyplot as plt

plt.plot(J_history)
plt.plot(w)
plt.plot(b)
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Training Cost vs. Iterations')
plt.show()
```

```python
def predict(x, w, b):
    return w * x + b

new_cities = [3.5, 7.0]  # Example populations in 10,000s
predictions = predict(np.array(new_cities), w, b)
print(f'Predicted profits: {predictions}')
```
